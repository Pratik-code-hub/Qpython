```python
# 🔹 1. No Parameters, No Return Value

def greet():
    print("Hello , pratik")
greet()
greet()
```

```python
# 🔹 2. Parameters, No Return Value

def greet(name):
    print("Hello",name) 
greet("Pratik")

```

```python
def add(a,b):
    print("sum is:", a+b)
add(3,5)
```

```python
# 🔹 3. Parameters with Return Value

def add(a,b):
    return a+b
result1 = add(4,6)

print(result1)
```

```python
def square(n):
    return n**2
result=square(5)
print(result)
```

```python
def square(n):
    return n*n
print(square(5))
```

```python
#🔹 4. No Parameters but Return Value
def get_number():
    return 10

print(get_number())
```

```python
def get_message():
    return "Python is easy!"

msg = get_message()
print(msg)
```
