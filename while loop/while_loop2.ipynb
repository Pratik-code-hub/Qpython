What is a Perfect Square?

A perfect square is a number obtained by multiplying a number by itself.

Examples:

1×1=1
2×2=4
3×3=9
4×4=16

So, numbers like:

1, 4, 9, 16, 25...

are perfect squares.

Perfect squares from 1 to 100
1, 4, 9, 16, 25, 36, 49, 64, 81, 100

```python
#Q22.)  Common Factor:  ( https://leetcode.com/problems/number-of-common-factors/description/ ) 2427

class Solution:
    def commonFactors(self, a: int, b: int) -> int:
        
        i = 1
        count = 0

        while i <= a and i <= b:
            if a % i == 0 and b % i == 0:
                count += 1

            i += 1

        return count
```

```python
# Q17.) 

class Solution:
    def subtractProductAndSum(self, n: int) -> int:
        product = 1
        summ = 0

        while n > 0:
            digit = n % 10

            product = product * digit
            summ = summ + digit

            n = n // 10

        return product - summ
    
```
