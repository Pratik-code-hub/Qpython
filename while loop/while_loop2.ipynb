```python
#Q22.)  Common Factor:  ( https://leetcode.com/problems/number-of-common-factors/description/ ) 2427

class Solution:
    def commonFactors(self, a: int, b: int) -> int:
        
        i = 1
        count = 0

        while i <= a and i <= b:
            if a % i == 0 and b % i == 0:
                count += 1

            i += 1

        return count
```

```python
# # Q17.) While Loop Digit based Programming
# Subtract the Product and Sum of Digits of an 
# # Integer        (https://leetcode.com/problems/subtract-the-product-and-sum-of-digits-of-an-integer/submissions/1995880017/)

class Solution:
    def subtractProductAndSum(self, n: int) -> int:
        product = 1
        summ = 0

        while n > 0:
            digit = n % 10

            product = product * digit
            summ = summ + digit

            n = n // 10

        return product - summ
    
```

```python
#Q18.)  Add Digits ( https://leetcode.com/problems/add-digits/submissions/1996760230/)

class Solution:
    def addDigits(self, num: int) -> int:

        while num >= 10:
            summ = 0

            while num > 0:
                digit = num % 10
                summ += digit
                num = num // 10

            num = summ

        return num

```

```python
 # Mirror distance of an integer. (https://leetcode.com/problems/mirror-distance-of-an-integer/submissions/1997187112/)
class Solution:
    def mirrorDistance(self, n: int) -> int:

        original = n
        rev = 0

        while n > 0:
            digit = n % 10
            rev = rev * 10 + digit
            n = n // 10

        return abs(original - rev)
```

```python
#  program to take a String input and 
# check whether the given String is a palindrome 
# String or not.
# Palindrome String: if String is same from both side 
# Input:
# String s1=“madam”; 
# Output:
# madam is a Palindrome String

s = "madam"

original = s
rev = ""

while len(s) > 0:
    rev = rev + s[len(s) - 1]
    s = s[:len(s) - 1]

if original == rev:
    print(original, "is a Palindrome String")
else:
    print(original, "is Not a Palindrome String")
```

```python
# https://leetcode.com/problems/number-of-steps-to-reduce-a-number-to-zero/description/

class Solution:
    def numberOfSteps(self, num: int) -> int:

        count = 0

        while num > 0:

            if num % 2 == 0:
                num = num // 2

            else:
                num = num - 1

            count += 1

        return count
```

```python
#  Factorial Trailing Zeros  ( https://leetcode.com/problems/factorial-trailing-zeroes/submissions/1997287247/  )

class Solution:
    def trailingZeroes(self, n: int) -> int:

        count = 0
        while n > 0:
            n //= 5
            count += n

        return count
```

```python
#  Power of Two     ( https://leetcode.com/problems/power-of-two/description/ )

class Solution:
    def isPowerOfTwo(self, n: int) -> bool:

        if n <= 0:
            return False

        while n % 2 == 0:
            n //= 2

        if n == 1:
            return True
        else:
            return False
```

```python
# Q28 Power of Three  (https://leetcode.com/problems/power-of-three/ )

class Solution:
    def isPowerOfThree(self, n: int) -> bool:
        
        if n<=0:
            return False
        while n%3==0:
            n//=3

        if n==1:
            return True
        else:
            return False
```

```python
# Q30.)  Number of 1 Bits:    ( https://leetcode.com/problems/number-of-1-bits/  )
class Solution:
    def hammingWeight(self, n: int) -> int:
        
        count=0
        while n>0:
            if n%2==1:
                count+=1
            n//=2
        return count
```

```python
#  to take binary input and print its decimal 
# representation.
# i/p: 10100
# o/p: 20

binary = 10100

decimal = 0
power = 0

while binary > 0:

    digit = binary % 10

    decimal = decimal + digit * (2 ** power)

    power += 1

    binary = binary // 10

print(decimal)

```
