What is a Perfect Square?

A perfect square is a number obtained by multiplying a number by itself.

Examples:

1×1=1
2×2=4
3×3=9
4×4=16

So, numbers like:

1, 4, 9, 16, 25...

are perfect squares.

Perfect squares from 1 to 100
1, 4, 9, 16, 25, 36, 49, 64, 81, 100

```python
# WAP to print all the numbers from 1 to 
# 100 which are perfect square.

n=100
i=1

while i*i<=n:
    print(i*i)
    i+=1
```

```python
# Q5.)  WAP to print and count all the numbers 
# from 1 to 100 which are perfect square.

n=100
i=1
count=0
while i*i<=n:
    count=count+1
    i+=1
print(count)
```

```python
# # Q6.)  bulb switcher   (https://leetcode.com/problems/bulb-switcher/description/)

# i=1
# count=0
# while i*i<=100:
#     count=count+1
#     i+=1
# print(count)

# 0r 


class Solution:
    def bulbSwitch(self, n: int) -> int:
        i = 1
        count = 0

        while i * i <= n:
            count += 1
            i += 1

        return count
```
