```python
# 49.) WAP to toggle a string.

s=input("enter a string")
result=""
i=0 
while i<len(s):
    if 'a'<= s[i] <='z':
        result+=chr(ord(s[i])-32)
    elif 'A'<= s[i]<= 'Z':
        result=result+chr(ord(s[i])+32)    
    else:
        result+=s[i]
    i+=1
print(result)
```

```python
# # WAP to reverse the given number (without typecasting)
n=int(input("enter a number")) #234
rev=0
while n>0:
    digit=n%10
    rev=rev*10+digit
    n=n//10

print(rev)
```

```python
# WAP to find the sum of individual digits of a number.
n=int(input('enter an number')) #4321 = (4+3+2+1)--> 120
sum=0
while n>0:
    digit=n%10
    sum=sum+digit
    n=n//10
print(sum)
```
