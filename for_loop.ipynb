```python
# 49.) WAP to toggle a string.
s=input("enter a string")
result=""
for ch in s:
    if  "a"<=ch<='z':
        result+=chr(ord(ch)-32)
    elif "A"<=ch<='Z':
        result+=chr(ord(ch)+32)
    else:
        result=result+ch
print(result)
```

```python
# # WAP to reverse the given number (without typecasting)
n=int(input("enter a number")) #234
rev=0
for i in str(n):
    digit=n%10
    rev=rev*10 + digit
    n=n//10
print(rev)
```

```python
# 58.) WAP to find the greatest number in a given list of integers. using indexing
lst = [2,45,2,2,4,5,676,6,33,22,9,488,90,24] 

max_num= lst[0] 
for i in range(len(lst)):
    if lst[i]> max_num:
        max_num= lst[i]
print(max_num)
```
