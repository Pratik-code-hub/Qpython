```python
#QUES 92.) WAP to get the following output. using for loop
#s=['jiocinema.com','file.py','web.html','amazon.com','www.org','python.py']
#out={'com':['jiocinema','amazon'],'py':['file','python'],'html':['web'],'org':['www']}

s = ['jiocinema.com', 'file.py', 'web.html', 'amazon.com', 'www.org', 'python.py']
out = {}
for i in s:
    parts = i.split('.')
    name = parts[0]
    e = parts[-1]
    if e not in out:
        out[e] = []
    out[e].append(name)
print(out)
```
